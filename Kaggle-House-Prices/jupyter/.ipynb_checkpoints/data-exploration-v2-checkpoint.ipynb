```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df_train = pd.read_csv('../dataset/train.csv')
df_test = pd.read_csv('../dataset/test.csv')
df = pd.concat([df_train, df_test])
# Restet index of df 
test_start_id = len(df_train)
test_start_id
```

```python
variable_list = list(df_train.columns)
variable_list
```

```python
num_variable = []
cate_variable = []
for i, k in enumerate(variable_list):
    if df_train[k].dtype == object:
        cate_variable.append(k)
    else:
        num_variable.append(k)
```

```python
# Numerical variable
```

```python
corr_matrix = df_train[num_variable].corr()
f, ax = plt.subplots(figsize=(10,8))
sns.heatmap(data=corr_matrix, vmax=0.8, square=True)
```

```python
# Sort corr_matrix
k = int(len(num_variable) * 0.5)
columns = corr_matrix.nlargest(k, 'SalePrice')['SalePrice'].index
corr_matrix_zoomed = df_train[columns].corr()
f, ax = plt.subplots(figsize=(15,10))
sns.heatmap(data=corr_matrix_zoomed, vmax=0.8, square=True, annot=True, yticklabels=columns.values, xticklabels=columns.values)
```

```python
num_variable = list(corr_matrix_zoomed[corr_matrix_zoomed['SalePrice']>0.5].index)
num_variable
```

```python
df[num_variable].isnull().any()
```

```python
df['GarageCars'].unique()
df['GarageArea'].unique()
df['TotalBsmtSF'].unique()
```

```python
df_train[num_variable].isnull().sum()
df_test[num_variable[1:]].isnull().sum()
df[num_variable].isnull().sum()
```

```python
df_test.ix[df_test['GarageCars'].isnull() == True ,'GarageCars'] = df_train['GarageCars'].mode()[0]
df_test.ix[df_test['GarageArea'].isnull() == True ,'GarageArea'] = df_train['GarageArea'].mean()
df_test.ix[df_test['TotalBsmtSF'].isnull() == True, 'TotalBsmtSF'] = df_train['TotalBsmtSF'].mean()
```

```python
df.iloc[test_start_id:, :] = df_test.loc[:,:]
```

```python
# Process Year
df['YearBuilt'] = 2017 - df['YearBuilt']
df['YearRemodAdd'] = 2017 - df['YearRemodAdd']
```

```python
df[num_variable]
```

```python
# Do log transform.
for i, k in enumerate(num_variable[1:]):
    stats.skew(df[k])
    if stats.skew(df[k]) > 0.5:
        print(k)
        df.loc[df[k] > 0, k] = np.log(df.loc[df[k] > 0, k])
```

```python
df.loc[df['SalePrice'].notnull() == True, 'SalePrice'] = np.log(df.loc[df['SalePrice'].notnull() == True, 'SalePrice'])
```

```python
drop_list = []
for i, k in enumerate(variable_list):
    if k not in num_variable:
        drop_list.append(k)
final_df = df.drop(drop_list, axis=1)
```

```python
# Category variable
```

```python
df = df.fillna('none')
for i, k in enumerate(cate_variable):
    temp = pd.get_dummies(df[k], prefix=k)
    final_df = pd.concat([final_df, temp], axis=1)
```

```python
final_df
```

```python
df_train = final_df.iloc[:test_start_id,:]
df_test = final_df.iloc[test_start_id:, :]
df_test = df_test.drop('SalePrice', axis=1)
with open('../dataset/train(preprocess_v2).csv', 'w') as file:
    df_train.to_csv(file, index=False)
with open('../dataset/test(preprocess_v2).csv', 'w') as file:
    df_test.to_csv(file, index=False)
```
