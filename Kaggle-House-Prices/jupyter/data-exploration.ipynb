```python
corr_matrix = df_train.corr()
f, ax = plt.subplots(figsize=(10,8))
sns.heatmap(data=corr_matrix, vmax=0.8, square=True)
```

```python
# Sort corr_matrix
k = 10
columns = corr_matrix.nlargest(k, 'SalePrice')['SalePrice'].index
corr_matrix_zoomed = np.corrcoef(df_train[columns].values.T)
f, ax = plt.subplots(figsize=(10,8))
sns.heatmap(data=corr_matrix_zoomed, vmax=0.8, square=True, annot=True, yticklabels=columns.values, xticklabels=columns.values)
```

```python
# Missing data
count = df_train.isnull().sum().sort_values(ascending=False)
percent = (df_train.isnull().sum() / len(df_train)).sort_values(ascending=False)
miss_data = pd.DataFrame({'Total': count, 'Percent':percent})
miss_data
```
