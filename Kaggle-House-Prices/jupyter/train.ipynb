```python
kf = KFold(n_splits=10, shuffle=True, random_state=123456)
```

```python
models = {}
for train_index, test_index in kf.split(x_train):
    xgb_regressor = xgboost.XGBRegressor()
    x = x_train.ix[train_index, :]
    y = y_train.ix[train_index]
    
    xgb_regressor.fit(x, y)
    y_predict = xgb_regressor.predict(x_train.ix[test_index,:])
    y_true = y_train.ix[test_index].values
    
    y_true = np.exp(y_true)
    y_predict = np.exp(y_predict)
    
    y_predict[0:10]
    y_true[0:10]
    
    error = np.linalg.norm(y_predict-y_true) / len(y_predict)
    input(error)
    models[error] = xgb_regressor
```

```python
model = sorted(models.items(), key=lambda x: x[0])[0][1]
```

```python
y_predict = model.predict(x_test)
```
